**WEB SCRAPPING THE CODEFORCES WEBSITE**

```python
#Importing the relevant libraries

from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Taking random codeforces handles

df_handles=pd.read_excel('C:/Users/hp/Downloads/Codeforces_Handles1.xlsx')
handles=df_handles['HANDLE'].tolist()
```

```python
data=[]
```

```python
#Extracting relevant information from each handle

for handle in handles:
    url_profile = f'https://codeforces.com/profile/{handle}'
    url_contests = f'https://codeforces.com/contests/with/{handle}'

    try:
        no_of_problems_solved = None
        no_of_problems_solved_last_month = None
        no_of_contests = None
        rating = None
        first_contest_date = None
        
        
        driver= webdriver.Chrome('')
        driver.get(url_profile)
        content_profile=driver.page_source
        soup_profile=BeautifulSoup(content_profile, features="lxml")
        
        divs_profile = soup_profile.find_all('div', class_='_UserActivityFrame_frame')

        for div in divs_profile:
            counter_value_divs = div.find_all('div', class_='_UserActivityFrame_counterValue')
            if len(counter_value_divs) > 0:
                no_of_problems_solved = counter_value_divs[0].text.strip()
                no_of_problems_solved_last_month = counter_value_divs[2].text.strip()
                break
                
        
        driver= webdriver.Chrome('')
        driver.get(url_contests)
        content_contests=driver.page_source
        soup_contests=BeautifulSoup(content_contests, features="lxml")

        contests_table = soup_contests.find('table', class_='tablesorter user-contests-table')
        rows = contests_table.find_all('tr')
        col_ratings_no = rows[1].find_all('td')
        col_date = rows[-1].find_all('td')
         
        rating = col_ratings_no[6].text
        no_of_contests = col_ratings_no[0].text
        first_contest_date = col_date[2].text

        
        print(handle, no_of_problems_solved, no_of_problems_solved_last_month, no_of_contests, rating, first_contest_date)

        data.append({
            'CF_Handle': handle,
            'Rating': rating,
            'Total problems': no_of_problems_solved,
            'Problems last month': no_of_problems_solved_last_month,
            'Contest given': no_of_contests,
            'First contest date': first_contest_date
        })

    except Exception as e:
        print(f"An error occurred for handle {handle}: {e}")
```

```python
#Storing the extracted data in pandas dataframe

df_data = pd.DataFrame(data)
```

```python
df_data.to_csv('codeforces_data.csv', index=False)
```
